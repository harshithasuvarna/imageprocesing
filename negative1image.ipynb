```python
# Negative of an image
import cv2
import numpy as np
img=cv2.imread('img.jpg')
print(img.dtype)
img_neg=255-img
cv2.imshow('negative of an image',img_neg)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
   
# Read an image
image = cv2.imread('img.jpg')
   
# Apply log transformation method
c = 255 / np.log(1 + np.max(image))
log_image = c * (np.log(image + 1))
   
# Specify the data type so that
# float value will be converted to int
log_image = np.array(log_image, dtype = np.uint8)
   
# Display both images
plt.imshow(image)
plt.show()
plt.imshow(log_image)
plt.show()
```

```python
#block extraction from an image
from skimage.io import imshow,imread
import matplotlib.pyplot as plt
img1=imread('img.jpg')
imshow(img1)
fig,ax=plt.subplots(1,3,figsize=(6,4),sharey=True)
ax[0].imshow(img1[:,0:150])
ax[0].set_title('first block')
ax[1].imshow(img1[:,400:600])
ax[1].set_title('second block')
ax[2].imshow(img1[:,600:800])
ax[2].set_title('third block')
```

```python
# mathematical operation on the image
#adding,subtracting,multipuling,dividing
import cv2
import numpy as np
img1=cv2.imread('img1.png')
img2=cv2.imread('img2.png')
add=img1+img2
sub=img1-img2
mul=img1*img2
div=img1/img2
plt.imshow(add)
plt.show()
plt.imshow(sub)
plt.show()
plt.imshow(mul)
plt.show()
plt.imshow(div)
plt.imshow()
```
